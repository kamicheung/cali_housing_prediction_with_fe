## Feature engineering

Lets create 3 new features that proved good level of variance that explains variability in house value: 
- ocean proximity by average house value feature
- house age by average house value feature
- income by average house value feature
- longitude and latitude as a feature because it can give us numerical representations of exact locations

```python
#ocean prox by avg house value, let's call this ocean_value
df['ocean_value']=df.groupby('ocean_proximity')['median_house_value'].transform(np.mean) 
df['age_value']=df.groupby('housing_median_age')['median_house_value'].transform(np.mean) #try median? dist is right skewed 
df['income_value']=df.groupby('median_income')['median_house_value'].transform(np.mean) #try median? dist is right skewed
```

```python
#lets label encode our ocean_proximity nominally
df['ocean_proximity']=df['ocean_proximity'].replace(['INLAND','<1H OCEAN','NEAR OCEAN','NEAR BAY','ISLAND'],[1,2,3,4,5])
```

```python
#longitude and latitude feature, let's call this location
df['location']=(df['longitude'].values*df['latitude'].values)*-1
```

```python
df
```

```python
sns.relplot(x='location',y='median_house_value',data=df,estimator=np.mean,hue='ocean_proximity')
```

There is a relationship between location value and house value, where 3800 to 4000 and 4400 to 4800 have high house value. So let's keep this new feature 'location' as it might help to explain the variability in the house value.

## Scaling & Model selection

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_dev=scaler.fit_transform(X_dev)
X_test=scaler.fit_transform(X_test)
X_dev
```

```python
def rmse(predictions, targets): 
  return np.sqrt(((predictions - targets) ** 2).mean())

rmse_score = make_scorer(rmse, greater_is_better=False)
```

**With feature engineering now, our models have improved r2 scores of more than 15%**

```python
#Linear Regression
model = LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_dev)
print(model.score(X_dev,y_actual),'rmse:',rmse(y_actual,y_pred))
```

```python
#Support Vector Regression
model = SVR(kernel='linear')
model.fit(X_train,y_train)
y_pred=model.predict(X_dev)
print(model.score(X_dev,y_actual),'rmse:',rmse(y_actual,y_pred))
```

```python
#Decision Tree Regressor
model = DecisionTreeRegressor(min_samples_leaf=100,max_depth=20,max_features='auto')
model.fit(X_train,y_train)
y_pred=model.predict(X_dev)
print(model.score(X_dev,y_actual),'rmse:',rmse(y_actual,y_pred))
```

```python
#Random Forest Regressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
y_pred=model.predict(X_dev)
print(model.score(X_dev,y_actual),'rmse:',rmse(y_actual,y_pred))
```

```python
#Light gradient boosting machine regressor
model = LGBMRegressor()
model.fit(X_train,y_train)
y_pred=model.predict(X_dev)
print(model.score(X_dev,y_actual),'rmse:',rmse(y_actual,y_pred))
```

Lightgbm is the best model, we should tune this to improve performance.

## Hypertuning

```python
#Initialize model with some parameters, 500 trees, 200 iterations, lr is a list of values
lgb = LGBMRegressor(metric = 'rmse',subsample = 0.5, num_iterations =200,  random_state=0 )
param_test ={'learning_rate' : [0.05,0.1,0.2,0.3],
            'num_leaves' : [500,800,1000],
            'reg_alpha' : [0.5,1.0,50.0,100.0],
            'reg_lambda' : [0.5,1.0,50.0,100.0]}
```

```python
#GridSearch over 100 iterations
Total_sets = 100

from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
gs = RandomizedSearchCV(
    estimator=lgb, param_distributions=param_test, 
    n_iter=Total_sets,
    scoring=rmse_score,
    cv=4,
    refit=True,
    random_state=314,
    n_jobs = 4,
    verbose=True)
gs.fit(X_train, y_train)

print('Best score reached: {} with params: {} '.format(-1*gs.best_score_, gs.best_params_))

y_pred  = gs.predict(X_dev)
dev_score =  rmse(y_pred,y_actual)
y_test_pred = gs.predict(X_test)
test_score =  rmse(y_test,y_test_pred)

print('dev r2 score:',r2_score(y_actual, y_pred),'dev rmse:',dev_score)
print('test r2 score:',r2_score(y_test, y_test_pred),'test rmse:',test_score)
```

Lightgbm tuning increased performance by 0.9%, and feature engineering increased performance by 17.5%. We could gather more data and feature engineer on more features mentioned previously to increase performance.
